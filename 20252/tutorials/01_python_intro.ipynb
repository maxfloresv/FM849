### Una estructura fundamental: lista de elementos

Las listas son una colecci√≥n ordenada de elementos, a diferencia de los conjuntos, donde el orden no importa. Esta estructura ser√° **fundamental** para este curso, pues el trabajo con datos se reduce a operaciones que pueden ser pensadas como listas.

En Python, una lista se define con corchetes $\texttt{[]}$, y los elementos separados por comas. Por ejemplo, la siguiente es la definici√≥n de la lista de n√∫meros $\texttt{L}$:

```python
L = [1, 2, 3, 4]
L
```

Esta estructura es **mutable**, es decir, su definici√≥n puede cambiar durante la ejecuci√≥n de las celdas de c√≥digo. Una de las operaciones fundamentales es el acceso mediante un √≠ndice. Se llamar√° √≠ndice a la posici√≥n de un elemento en la lista. Las listas est√°n indexadas desde $0$. As√≠, $\texttt{L[0]}$ ser√° el primer elemento de $\texttt{L}$, y $\texttt{L[n-1]}$ el √∫ltimo, si se considera que la lista tiene $n$ elementos.

En el siguiente celda de c√≥digo, se accede a las posiciones $0$ y $n-1$ de la lista $\texttt{L}$ creada anteriormente. El valor de $n$ se puede determinar usando la funci√≥n `len()`, que sirve para cualquier lista.


```python
n = len(L)
print("El primer elemento de la lista es:", L[0])
print("El √∫ltimo elemento de la lista es:", L[n-1])
```

> ‚ùó **Importante**: Un error muy com√∫n al trabajar con listas es olvidar que el √≠ndice parte desde $0$. Por ejemplo, ¬°$\texttt{L[n]}$ arroja error (`list index out of range`)!

Previamente, se hab√≠a mencionado que las listas son mutables. Se comprobar√° con la definici√≥n de $\texttt{L}$, notando que la siguiente es una celda de c√≥digo independiente de las anteriores.

```python
print("Definici√≥n de L original:", L)
# Se cambia el elemento en el √≠ndice 2 por -1
L[2] = -1
print("Mutaci√≥n de L dada la instrucci√≥n:", L)
```

Como ya se vio la definici√≥n inicial de una lista, ahora se ver√°n sus principales m√©todos en acci√≥n.

- `append()` - Agrega un elemento al final de la lista.
- `extend()` - Agrega los elementos de otra lista (o estructura que se pueda recorrer, como por ejemplo, un _string_) al final de la lista.
- `pop()` - Elimina el elemento de la lista en una determinada posici√≥n (si no se pasa la posici√≥n como argumento, por defecto es el √∫ltimo).
- `sort()` - Ordena la lista seg√∫n un criterio (si no se pasa el argumento `reverse`, por defecto se ordena de menor a mayor).

```python
print("Versi√≥n original de L:", L)

# Se agrega el 3 al final de L
L.append(3)
print("Elemento 3 agregado al final:", L)

L2 = [8, 3, 4]
# Se agrega L2 al final de L
L.extend(L2)
print("Extensi√≥n con L2:", L)

# Se elimina el √∫ltimo elemento de L
L.pop()
print("√öltimo elemento eliminado:", L)

# Se elimina el elemento de L en el √≠ndice 2
L.pop(2)
print("Elemento en √≠ndice 2 eliminado:", L)

# Ordena de menor a mayor
L.sort()
print("L ordenado ascendentemente:", L)

# Ordena de mayor a menor
L.sort(reverse=True)
print("L ordenado descendentemente:", L)
```

> ‚ùó **Importante**: Al eliminar un elemento con `pop()`, previamente se debe asegurar de que dicho √≠ndice exista, sino, el programa lanzar√° un error.

### Diccionarios

Un diccionario es una estructura que guarda pares de elementos, donde uno se asocia a otro. El elemento representante del par se llama llave o clave, y el elemento guardado en la asociaci√≥n se llama valor. En Python, se definen con llaves $\texttt{\{\}}$. La clave se busca que sea √∫nica, pues en caso de que se repita, se sobreescribe el valor anterior guardado.

Un ejemplo de uso es cuando se busca guardar el nombre de cada persona dado su RUT. Ac√°, la clave es el RUT, porque corresponde a un identificador. Si se usara el nombre como clave y el RUT como valor, existe una probabilidad de que dos personas tengan el mismo nombre, y por lo tanto, se generar√≠a una sobreescritura del RUT.

Para guardar un valor en un diccionario, se usa la instrucci√≥n $\texttt{D[k] = v}$, donde $\texttt{D}$ es el nombre del diccionario, $\texttt{k}$ la clave, y $\texttt{v}$ el valor. Esto se puede observar en la siguiente celda de c√≥digo:

```python
nombres = {}

# 18.293.731 es el RUT de Juan
nombres["18.293.731"] = "Juan"
nombres["19.116.224"] = "Pedro"

# Se pueden usar otros tipos de datos para las claves
nombres[19007211] = "Diego"

nombres
```

```python
nombres_2 = {
    # llave: valor
    "18.293.731": "Juan",
    "19.116.224": "Pedro",
    19007211: "Diego"
}

nombres_2
```

A continuaci√≥n, se observa un ejemplo de sobreescritura cuando la clave no es √∫nica. El √∫ltimo valor definido para la clave es el que queda, en este caso, $\texttt{Texto 2}$.

```python
D = {}

D[0] = "Texto 1"
print("El diccionario es", D)
D[0] = "Texto 2"

D
```

#### Condicionales

Los condicionales son instrucciones de c√≥digo que permiten separar la l√≥gica cuando una cierta condici√≥n se cumple y cuando no se cumple. Se anotan con la palabra clave $\texttt{if}$. Un ejemplo es el siguiente, donde se quiere imprimir ¬´$x$ es mayor que $6$¬ª cuando la condici√≥n $x > 6$ se cumple, y ¬´$x$ es menor o igual que $6$¬ª cuando no. La palabra $\texttt{else}$ se usa para indicar que lo que est√© dentro se debe ejecutar cuando la condici√≥n no se cumple.

```python
x = 3

if x > 6:
  print("x es mayor que 6")
else:
  print("x es menor o igual que 6")
```

Se debe notar que el c√≥digo ahora tiene unos espacios despu√©s de las instrucciones $\texttt{if}$ y $\texttt{else}$. Estos espacios se llaman indentaciones, se insertan a trav√©s de la tecla $\texttt{Tab}$, definen jerarqu√≠as, y son esenciales para que el c√≥digo sea interpretado correctamente. El uso de condicionales introduce jerarqu√≠a, porque el c√≥digo que sigue est√° ¬´envuelto¬ª seg√∫n si se cumple la condici√≥n o no.

Por otro lado, algunas de las comparaciones b√°sicas que se pueden hacer con n√∫meros son las siguientes:

- `>`, `<` - Desigualdad estricta (mayor que, menor que).
- `==` - Igualdad (‚ùó Ojo, a diferencia de la asignaci√≥n de variables, para la comparaci√≥n se usan dos signos `=`).
- `>=`, `<=` - Desigualdad no estricta (mayor o igual que, menor o igual que).

Cuando se quieren introducir varias condiciones, por ejemplo, clasificar a una persona seg√∫n su edad (menor de edad, adulto, adulto mayor), se usa la instrucci√≥n $\texttt{elif}$. La ejecuci√≥n siempre se hace de arriba abajo, por lo tanto, si alguna condici√≥n se cumple, no se ejecutar√° ninguna m√°s de abajo cuando se ocupe esta variante.

```python
edad = 7

if edad < 18:
  print("Menor de edad")
# No es necesario poner edad >= 18.
elif edad < 60:
  print("Adulto")
else: # >= 60
  print("Adulto mayor")
```

Esto se debe interpretar as√≠: ¬´si la edad de la persona es menor estricta que $18$, marcar como menor de edad. Sino, si la edad de la persona es menor estricta que $60$ (se asume que es mayor o igual que $18$ porque la condici√≥n anterior no se cumpli√≥, y se ejecuta de arriba abajo), marcar como adulto. Sino, marcar como adulto mayor¬ª. La √∫nica manera de que haya alcanzado la instrucci√≥n $\texttt{else}$ es que ninguna de las condiciones anteriores se haya cumplido.

#### Expresiones compuestas por _booleanos_

A veces, existe la necesidad de pedir que se cumplan dos o m√°s condiciones en simult√°neo, o alguna de ellas, o ninguna. Para ello, se introducen los conectivos l√≥gicos $\texttt{and}$, $\texttt{or}$ y $\texttt{not}$. Cada uno de ellos significa ¬´y¬ª, ¬´o¬ª y ¬´no¬ª (de negaci√≥n) respectivamente.

```python
edad = 74
nombre = "Pedro"

print("Prueba con and:", edad == 74 and nombre == "Juan")
print("Prueba con or:", edad == 74 or nombre == "Juan")
print("Prueba con not:", not edad == 74)
```

La primera instrucci√≥n es $\texttt{False}$, porque $\texttt{and}$ exige que ambas condiciones se cumplan en simult√°neo, y el nombre no es $\texttt{Juan}$. La segunda es $\texttt{True}$, porque $\texttt{or}$ exige que al menos una sea verdad, y la edad s√≠ es $74$. Por √∫ltimo, la tercera es $\texttt{False}$, porque la edad s√≠ es $74$, entonces se est√° negando una verdad.

### Funciones

Las funciones permiten envolver caracter√≠sticas de c√≥digo reutilizables a trav√©s de un nombre. √âstas se definen con la palabra clave $\texttt{def}$ acompa√±ada de su nombre. Cuentan con par√°metros, que son variables que se procesar√°n dentro del bloque de la funci√≥n. Los par√°metros se definen entre par√©ntesis $\texttt{(}\dots\texttt{)}$ inmediatamente despu√©s del nombre de la funci√≥n (ver ejemplo de abajo). Para que la funci√≥n devuelva el resultado de una o varias operaciones, hay que escribir la instrucci√≥n $\texttt{return}$ al final de ella.

Por ejemplo, la siguiente funci√≥n se llama $\texttt{func}$, y recibe los par√°metros $\texttt{a}$ y $\texttt{b}$, devolviendo su suma.

```python
def func(a, b):
  return a + b

func(1, 3)
```

Como se puede ver, una funci√≥n se invoca por su nombre, y en los par√°metros que acepta se le pasan los valores que procesa. En el caso anterior, $a=1$ y $b=3$, porque los par√°metros se procesan en el mismo orden en el que se definen.

Una ventaja de las funciones es que permiten definir par√°metros por omisi√≥n. Los par√°metros por omisi√≥n siempre se definen de los √∫ltimos, y deben tener un valor por defecto, que se asigna de la misma forma que se asignan variables, como se puede ver en el siguiente bloque de c√≥digo:

```python
def func_2(a, omision=1):
  return a * omision

print("func_2 sin pasar el par√°metro por omisi√≥n:", func_2(4))
print("func_2 explicitando el par√°m. por omisi√≥n:", func_2(4, 2))
```

La explicaci√≥n de este c√≥digo es la siguiente: cuando no se pasa el segundo par√°metro, se usa el valor por defecto del par√°metro por omisi√≥n, es decir, $1$, entonces el resultado es $4 \cdot 1 = 4$. En el segundo caso, $\texttt{omision} = 2$, por lo tanto, el resultado es $4 \cdot 2 = 8$.

Como este tipo de funciones devuelven valores mediante la instrucci√≥n $\texttt{return}$, se pueden asignar a una variable. Por ejemplo:

```python
val = func_2(8)
val + 2
```

La explicaci√≥n es la misma: $\texttt{func_2(8)}$ devuelve $8$, porque se usa el valor por defecto del par√°metro por omisi√≥n ($1$), por lo tanto, $\texttt{val} = 8$ y $\texttt{val} + 2 = 10$, que es el valor devuelto.

En los dos casos anteriores, las funciones ten√≠an s√≥lo dos par√°metros. Esto no tiene por qu√© ser as√≠, pueden tener cero o m√°s. Lo √∫nico necesario es que los par√°metros por omisi√≥n siempre est√©n definidos al final con un valor por defecto.

Si una funci√≥n no incluye un valor de retorno mediante $\texttt{return}$, entonces no devuelve nada. Esto se puede comprobar en el siguiente bloque de c√≥digo:

```python
def func_3(a):
  b = a + 1

c = func_3(2)
print(c)
```

La funci√≥n nunca ocupa la instrucci√≥n $\texttt{return}$, por lo tanto, el valor retornado es $\texttt{None}$.

> ‚ùó Ojo, una confusi√≥n com√∫n es pensar que una funci√≥n ¬´devuelve¬ª algo cuando se ocupa `print()` dentro de ella. Esto no es cierto, porque `print()` siempre muestra un mensaje en pantalla. Se puede comprobar con el extracto de c√≥digo de abajo.

```python
def func_4():
  print("Esto no es un mensaje de retorno")

c = func_4()
print(c)
```

El c√≥digo deber√≠a interpretarse de la siguiente forma: la funci√≥n `func_4()` imprime en pantalla el mensaje ¬´Esto no es un mensaje de retorno¬ª, entonces cuando se instruye $\texttt{c} = \texttt{func_4()}$, la funci√≥n est√° siendo llamada, mostrando el mensaje anterior en pantalla. Sin embargo, al final se escribe $\texttt{print(c)}$, pero `func_4()` no devuelve nada, por lo tanto, $\texttt{c}$ es $\texttt{None}$.

Esta funci√≥n se envuelve entre par√©ntesis $\texttt{(lambda} \dots \texttt{)}$ para ser llamada inmediatamente con los par√°metros $\texttt{x} = 6$ e $\texttt{y} = 3$, devolviendo el resultado $6-3=3$. M√°s adelante en el curso se mostrar√° una aplicaci√≥n de las funciones an√≥nimas en el contexto de Ciencia de Datos.

### Ciclos

La √∫ltima componente principal que se revisar√° en este _notebook_ antes de pasar al trabajo con datos ser√°n los ciclos. Un ciclo permite repetir un fragmento de c√≥digo las veces que se requiera, incluso infinitamente. A prop√≥sito de ciclos infinitos, la forma de forzar la ruptura de un ciclo es mediante la instrucci√≥n $\texttt{break}$.

Hay dos formas de definir un ciclo: la primera es mediante la instrucci√≥n $\texttt{for} \dots \texttt{in}$, y la segunda mediante la instrucci√≥n $\texttt{while}$.

#### Ciclos $\texttt{for}$

Se usan principalmente para definir ciclos que se mueven dentro de un rango definido, por ejemplo, para moverse en el rango de n√∫meros $\{1, 2, \dots, 10\}$, y repetir instrucciones de c√≥digo movi√©ndose en orden en cada uno de dichos n√∫meros (llamados tambi√©n puntos de iteraci√≥n).

Por convenci√≥n, la variable que se usa para definir cu√°l es la iteraci√≥n actual es $\texttt{i}$ (de _index_, √≠ndice). Cuando $\texttt{i}$ est√° ocupado, se suele usar $\texttt{j}$, $\texttt{k}$, y as√≠. Para aclarar este concepto, a continuaci√≥n hay un bloque de c√≥digo que lo implementa.

```python
for i in range(1, 11):
  print(i)
```

Como se puede observar, el c√≥digo anterior imprime todos los n√∫meros del $1$ al $10$. La explicaci√≥n es la siguiente: $\texttt{range(a, b)}$ define un rango de n√∫meros enteros en el intervalo $[a, b)$ (es decir, excluyendo a $b$). Por eso, $\texttt{range(1, 11)}$ parte en $1$ y llega a $10$.

La funci√≥n `range()` tambi√©n se puede invocar de las siguientes formas:

- Un √∫nico par√°metro ($\texttt{range(a)}$): se mueve en el rango de n√∫meros enteros $[0, a)$.

- Usar tres par√°metros ($\texttt{range(a, b, s)}$): se mueve en el rango de n√∫meros enteros $[a, b)$, pero dando saltos de tama√±o $s$. Si no se especifica $s$, por defecto es $1$, por eso los n√∫meros enteros crecen de $1$ en $1$.

```python
for i in range(3):
  print("S√≥lo range(3)", i)

for i in range(1, 15, 2):
  print("Con step=2", i)
```

> ‚ùó Ojo, que al definir un ciclo $\texttt{for}$, la variable definida para marcar las iteraciones (en este caso $\texttt{i}$) s√≥lo existe en el bloque de c√≥digo que est√° dentro del ciclo.

El uso m√°s com√∫n cuando se trabaja con datos es recorrer toda una lista. Previamente, se revis√≥ que `len()` devolv√≠a al cantidad de elementos dentro de una lista, y adem√°s, $\texttt{L[i]}$ devolv√≠a el elemento en el √≠ndice $i$. El conocimiento se puede mezclar:

```python
L = [1, 3, 5, 8]
for i in range(len(L)):
  print(L[i])
```

Este c√≥digo puede ser un poco m√°s complejo de entender, pero hay que interpretarlo de la siguiente forma: $\texttt{L}$ es una lista con $4$ elementos, por lo tanto, $\texttt{len(L)} = 4$. As√≠, lo de abajo se lee `for i in range(4)`. Ya se revis√≥ que en este caso, $i$ recorre los n√∫meros enteros en el intervalo $[0, 4)$. As√≠, en cada iteraci√≥n, se imprime $\texttt{L[i]} = \texttt{L[0]}$ (primera iteraci√≥n), $\texttt{L[1]}$ (segunda iteraci√≥n), $\texttt{L[2]}$ (tercera iteraci√≥n) y $\texttt{L[3]}$ (√∫ltima iteraci√≥n). Por eso, se imprimen todos los elementos de $\texttt{L}$.

#### Ciclos $\texttt{while}$



A diferencia de los ciclos $\texttt{for}$, los ciclos $\texttt{while}$ se usan en el caso general cuando se busca que una condici√≥n se cumpla. Por ejemplo, se quiere repetir un c√≥digo mientras un n√∫mero $a$ es menor que $b$, entonces se puede implementar de la siguiente manera:

```python
a = -3
b = 8
while a < b:
  print("a es menor que b:", a, b)
  # a += 1 se ocupa para sumarle 1 unidad a la variable "a"
  a += 1
```

Como se puede observar, cada vez que se ejecuta el ciclo, se le suma una unidad a la variable $a$ mediante la instrucci√≥n $\texttt{+= 1}$. Eventualmente, $a$ alcanzar√° a $b$. Esto pasa en la iteraci√≥n cuando $a=7$, pues se vuelve a sumar $1$ y el programa queda en el estado $a=8$ y $b=8$. Finalmente, se verifica si la condici√≥n $a < b$ se cumple, y como no es as√≠, el ciclo se deja de ejecutar. De esta manera, el √∫ltimo mensaje en pantalla que se ve es ¬´$a$ es menor que $b$: $7$ $8$¬ª.

Algo importante de mencionar es que $a$ debe aumentar en una unidad en este contexto, sino nunca podr√° alcanzar a $b$ y el ciclo se ejecutar√° infinitamente. Tambi√©n, si $a \ge b$ en la iteraci√≥n inicial (por ejemplo, si hubi√©semos definido al principio $a=8$ y $b=7$), nunca hubiese entrado al ciclo, porque inicialmente la condici√≥n no era verdadera.

Finalmente, la manera de romper un ciclo en cualquier caso, como se mencion√≥ anteriormente, es mediante la instrucci√≥n $\texttt{break}$. Esto funciona tanto para los $\texttt{for}$ como para los $\texttt{while}$. A continuaci√≥n se muestra un ejemplo que mezcla varios aprendizajes previos:

```python
s = "Esto es una prueba"
i = 0
while s[i] != "p":
  if s[i] == " ":
    break
  print("Caracter actual:", s[i])
  i += 1
```

Este c√≥digo tambi√©n es complejo de entender en primera instancia, pero se le debe dar una vuelta. La interpretaci√≥n es la siguiente: $\texttt{s}$ es el texto ¬´Esto es una prueba¬ª. $\texttt{i}$ refleja un √≠ndice que parte de $0$ (parecido a lo que se hac√≠a con los ciclos $\texttt{for}$). La condici√≥n para que el ciclo se ejecute es que $\texttt{s[i]}$ (el caracter en el √≠ndice $i$ del texto $\texttt{s}$) sea distinto de ¬´p¬ª. Se sabe que es ¬´p¬ª cuando se alcanza la primera letra de ¬´prueba¬ª, pues es la primera letra coincidente de izquierda a derecha. Adem√°s, dentro del ciclo, se instruye que si el caracter actual $\texttt{s[i]}$ es un espacio, entonces debe romper tambi√©n el ciclo. As√≠, lo que se imprime son todas las letras hasta el primer espacio, que corresponden a las letras de la palabra ¬´Esto¬ª.
